```python

import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv("train.csv")

a = data["GrLivArea"]
b = data["OverallQual"]
c = data["SalePrice"]

a = a.fillna(0)
b = b.fillna(0)
c = c.fillna(0)

x = pd.DataFrame()
x["area"] = a
x["qual"] = b

model = LinearRegression()
model.fit(x, c)

p = model.predict(x)

print("first prices")
print(p[:10])

print("done")
```
