```python
class SimpleProlog:
    def __init__(self):
        self.facts = set()
        self.rules = []

    def add_fact(self, fact):
        self.facts.add(fact)

    def add_rule(self, head, body):
        self.rules.append((head, body))

    def ask(self, goal):
        if goal in self.facts:
            return True

        for head, body in self.rules:
            if head == goal:
                if all(self.ask(cond) for cond in body):
                    return True
        return False

print("Simple Prolog Engine Started")
```

```python
kb = SimpleProlog()

kb.add_fact(('parent', 'john', 'mike'))
kb.add_fact(('parent', 'john', 'sara'))
kb.add_fact(('parent', 'mike', 'tom'))

print("Stored Facts:")
for f in kb.facts:
    print(f"{f[0]}({f[1]}, {f[2]})")
```

```python
class HealthExpertSystem:

    def __init__(self):

        self.patient_symptoms = {}

        self.disease_db = {
            'cold': ['cough', 'sneezing', 'fever'],
            'malaria': ['fever', 'chills', 'headache'],
            'asthma': ['shortness_of_breath', 'cough']
        }

        self.medicine = {
            'cold': 'Rest and drink warm fluids',
            'malaria': 'Take anti-malarial medicine',
            'asthma': 'Use inhaler and consult doctor'
        }

    def add_symptom(self, patient, symptom):

        if patient not in self.patient_symptoms:
            self.patient_symptoms[patient] = []

        self.patient_symptoms[patient].append(symptom)

    def diagnose(self, patient):

        if patient not in self.patient_symptoms:
            return "unknown"

        ps = set(self.patient_symptoms[patient])

        for disease, sym in self.disease_db.items():
            if all(s in ps for s in sym):
                return disease

        return "unknown"

    def treatment(self, disease):

        return self.medicine.get(disease, "Visit hospital")
```

```python
sys = HealthExpertSystem()

sys.add_symptom("Ali", "cough")
sys.add_symptom("Ali", "sneezing")
sys.add_symptom("Ali", "fever")

sys.add_symptom("Ahmed", "shortness_of_breath")
sys.add_symptom("Ahmed", "cough")

for p in ["Ali", "Ahmed"]:

    d = sys.diagnose(p)
    t = sys.treatment(d)

    print("\nPatient:", p)
    print("Diagnosis:", d)
    print("Treatment:", t)
```

```python
class GrammarSystem:

    def __init__(self):

        self.det = {'the','a'}
        self.nouns = {'cat','dog','boy','apple'}
        self.verbs = {'eats','likes','sees'}

    def is_det(self,w):
        return w in self.det

    def is_noun(self,w):
        return w in self.nouns

    def is_verb(self,w):
        return w in self.verbs

    def check(self,sentence):

        words = sentence.split()

        if len(words) < 2:
            return False

        if words[0] in self.det:
            if len(words) >=3 and words[1] in self.nouns and words[2] in self.verbs:
                return True

        if words[0] in self.nouns and words[1] in self.verbs:
            return True

        return False


g = GrammarSystem()

tests = [
    "the cat eats",
    "dog eats",
    "cat dog",
    "the apple likes"
]

for s in tests:
    print(s, "->", g.check(s))
```

```python
class QA:

    def __init__(self):

        self.types = {
            'cat':'animal',
            'dog':'animal',
            'animal':'living'
        }

    def is_type(self,obj,typ):

        while obj in self.types:
            if self.types[obj] == typ:
                return True
            obj = self.types[obj]

        return False


qa = QA()

print("Is cat animal ?", qa.is_type("cat","animal"))
print("Is dog living ?", qa.is_type("dog","living"))
print("Is cat bird ?", qa.is_type("cat","bird"))
```
